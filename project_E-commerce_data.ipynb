TIME DATA ANALYSIS

```python
# CONVERT INVOICEDATE TO DATETIME
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# SET THE INVOICEDATE AS THE INDEX
df.set_index('InvoiceDate', inplace=True)

# GROUP BY MONTH AND SUM THE SALES
monthly_sales = df['Sales'].resample('M').sum()

# CREATE THE SALES GRAPH
plt.figure(figsize=(8, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum()

# PLOT MONTHLY SALES TRENDS
plt.figure(figsize=(18, 6))
monthly_sales.plot(kind='bar')
plt.title('Monthly Sales Trends')
plt.xlabel('Year, Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

THIS DATA SHOWS VARIABILITY ACROSS DIFFERENT MONTHS AND YEARS, WITH SPECIFIC MONTHS SUCH AS JANUARY AND DECEMBER 
AND YEAR SUCH AS 2019,2022 AND 2023 SHOWING HIGHER SALES.
AVERAGE SESSION DURATIONS ARE CONSISTENT, INDICATING STABLE CUSTOMER INTEREST.

```python
#monthly sales

df_2019 = df[df['Year'] == 2019]
monthly_sales_2019 = df_2019.groupby(['Month'])['Sales'].sum()

df_2020=df[df['Year']==2020]
monthly_sales_2020=df_2020.groupby(['Month'])['Sales'].sum()

df_2021=df[df['Year']==2021]
monthly_sales_2021=df_2021.groupby(['Month'])['Sales'].sum()

df_2022=df[df['Year']==2022]
monthly_sales_2022=df_2022.groupby(['Month'])['Sales'].sum()

df_2023=df[df['Year']==2023]
monthly_sales_2023=df_2023.groupby(['Month'])['Sales'].sum()

```

```python
plt.figure(figsize=(18, 10))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

plt.subplot(2,3,1)
monthly_sales_2019.plot(kind='bar', color='black')
plt.title('Sales in 2019')
plt.xlabel('month')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.subplot(2,3,2)
monthly_sales_2020.plot(kind='bar', color='green')
plt.title('Sales in 2020')
plt.xlabel('month')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.subplot(2,3,3)
monthly_sales_2021.plot(kind='bar', color='hotpink')
plt.title('Sales in 2021')
plt.xlabel('month')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.subplot(2,3,4)
monthly_sales_2022.plot(kind='bar', color='blue')
plt.title('Sales in 2022')
plt.xlabel('month')
plt.ylabel('Sales')
plt.xticks(rotation=0)

plt.subplot(2,3,5)
monthly_sales_2023.plot(kind='bar', color='red')
plt.title('Sales in 2023')
plt.xlabel('month')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
#AGGREGATE SALES BY YEAR AND CALCULATE THE TOTAL SALES PER YEAR 
yearly_sales = df.groupby('Year')['Sales'].sum()
print("Yearly Sales:")
print(yearly_sales)

# VISUALIZE YEARLY SALES TRENDS
plt.figure(figsize=(8, 4))
yearly_sales.plot(kind='bar', color='black')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
# EXTRACT MONTH AND YEAR FOR ANALYSIS
df['Month'] = df['InvoiceDate'].dt.month
df['Year'] = df['InvoiceDate'].dt.year

# CREATE A PIVOT TABLE TO SUMMARIZE SALES BY YEAR AND MONTH
pivot_table = df.pivot_table(values='Sales', index='Year', columns='Month', aggfunc='sum')

print("\nSales comparison by year and month:")
print(pivot_table)

# VISUALIZE THE DATA USING A HEATMAP 
plt.figure(figsize=(15, 10))
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="Greens", linewidths=.5)
plt.title('Sales Heatmap by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```
